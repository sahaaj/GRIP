***In this regression task we will predict the percentage of marks that a student is expected to score based upon the number of hours they studied. This is a simple linear regression task as it involves just two variables.***

# The libraries required to perform the task

```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.linear_model import LinearRegression
```

# Importing the data

```
dataset=pd.read_csv('http://bit.ly/w-data')
dataset.head()
```

Converting the data from the pandas dataframe into a numpy array so that we can use the data for training

```
X=dataset.iloc[:,0].values
X=X.reshape(-1,1)
Y=dataset.iloc[:,1].values
```

# Performing the Train Test Split

```
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.15)
```

# Performing the simple linear regression on the training data

```
linear_regressor=LinearRegression()
linear_regressor.fit(X_train,Y_train)
```

# Plotting out the results for better visualization

```
train_pred=linear_regressor.predict(X_train)
plt.scatter(X_train,Y_train,c="red")
plt.plot(X_train,train_pred,c="blue")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.title("Training plot")
plt.show()
```

# Making predictions

```
test_pred=linear_regressor.predict(X_test)
print("Actual Value \t Predictions")
for x,y in zip(test_pred,Y_test):
  print(y,"\t\t",x)
```

```
plt.scatter(X_test,Y_test,c="red")
plt.plot(X_test,test_pred,c="blue")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.title("Testing plot")
plt.show()
```

# Evaluating our model

```
train_error=metrics.mean_absolute_error(Y_train,train_pred)
test_error=metrics.mean_absolute_error(Y_test,test_pred)
print("Training error:",train_error)
print("Testing error:",test_error)
```

# What will be predicted score if a student study for 9.25 hrs in a day?

```
value=np.array([9.25]).reshape(-1,1)
output=linear_regressor.predict(value)
print(output)
```

```
print("If a student study for",value.squeeze(),"hours a day, then the predicted score will be",output[0])
```
